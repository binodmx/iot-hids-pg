```python
import os
import pandas as pd

df = pd.read_csv(os.getcwd() + '/../data/x-iiotid/X-IIoTID dataset.csv')
```

```python
print("No of Featueres: ", len(df.columns))
print("Features: ", df.columns)
```

```python
print("No of classes: ", len(df['class1'].unique()))
print(df['class1'].unique())

print("No of classes: ", len(df['class2'].unique()))
print(df['class2'].unique())

print("No of classes: ", len(df['class3'].unique()))
print(df['class3'].unique())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 6))
plt.xticks(rotation=90)
s = sns.countplot(data=df, x='class3', hue='class1')
for p in s.patches:
    s.annotate(format(p.get_height(), '.0f'), 
               (p.get_x() + p.get_width() / 2., p.get_height()), 
               ha = 'center', va = 'center', 
               xytext = (0, 9), 
               textcoords = 'offset points')
plt.title('Record Count', fontsize=20)

plt.show()
```

```python
# remove duplicate rows
df_no_duplicates = df.drop_duplicates()
# print length of dataset after removing duplicates
print("Length of dataset before removing duplicates: ", len(df))
print("Length of dataset after removing duplicates: ", len(df_no_duplicates))
```
