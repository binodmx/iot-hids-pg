```python
import matplotlib.pyplot as plt

color1 = (19/255, 130/255, 155/255)
color2 = (255/255, 180/255, 0/255)
color3 = (255/255, 97/255, 56/255)
color4 = (102/255, 205/255, 170/255)

x_ticks = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
attack = [1, 1, 1, 1]
benign = [0.63, 1, 0.63, 0.78]

bar_width = 0.25

plt.figure(figsize=(8,6))
plt.bar([i - 0.5 * bar_width for i in range(len(x_ticks))], attack, width=bar_width, color=color3, label='Attack')
plt.bar([i + 0.5 * bar_width for i in range(len(x_ticks))], benign, width=bar_width, color=color4, label='Benign')

plt.xticks([i for i in range(len(x_ticks))], x_ticks)

# plt.xlabel('Metric', fontdict={'fontweight': 'bold', 'fontsize': 32})
# plt.ylabel('', fontdict={'fontweight': 'bold', 'fontsize': 32})

plt.grid(which='major')
plt.legend(fontsize=24)

plt.rcParams['font.family'] = 'times new roman'
plt.tick_params(axis='both', labelsize=24)

plt.savefig("../charts/eval-vector-store-retrieval.png", format="png", bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt

color1 = (19/255, 130/255, 155/255)
color2 = (255/255, 180/255, 0/255)

x = ['Attack', 'Benign']
train_set_size = [39517, 19441]
test_set_size = [9879, 4860]

plt.figure(figsize=(8,6))

plt.bar(x, train_set_size, color=color1, width=0.25, label='Train')
plt.bar(x, test_set_size, color=color2, width=0.25, bottom=train_set_size, label='Test')

plt.ylim([0, 50000])

plt.grid(which='major')
plt.legend(fontsize=24)

plt.rcParams['font.family'] = 'times new roman'
plt.tick_params(axis='both', labelsize=24)


plt.savefig("../charts/dataset-size.png", format="png", bbox_inches="tight")
plt.show()
```
