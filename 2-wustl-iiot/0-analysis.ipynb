```python
import os
import pandas as pd

df = pd.read_csv(os.getcwd() + '/../data/wustl-iiot/wustl_iiot_2021.csv')
```

```python
print("No of Featueres: ", len(df.columns))
print("Features: ", df.columns)
```

```python
print("No of classes: ", len(df['Traffic'].unique()))
print(df['Traffic'].unique())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 6))
plt.xticks(rotation=90)
s = sns.countplot(data=df, x='Target', hue='Traffic')
for p in s.patches:
    s.annotate(format(p.get_height(), '.0f'), 
               (p.get_x() + p.get_width() / 2., p.get_height()), 
               ha = 'center', va = 'center', 
               xytext = (0, 9), 
               textcoords = 'offset points')
plt.title('Record Count', fontsize=20)

plt.show()
```
