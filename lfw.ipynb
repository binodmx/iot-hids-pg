```python
import os

dataset_path = "data/lfw/lfw-a"

# Count the number of people in the dataset
print("Total number of people: ", len(os.listdir(dataset_path)))

image_count = {}
for person in os.listdir(dataset_path):
    image_count[person] = len(os.listdir(os.path.join(dataset_path, person)))


# draw a histogram of the number of images per person
import matplotlib.pyplot as plt
plt.hist(image_count.values(), bins=range(1, 11))
plt.xlabel("Number of images")
plt.ylabel("Number of people")
plt.title("Number of images per person")
plt.show()
```
