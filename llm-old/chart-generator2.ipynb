```python
a = {'DDoS_HTTP': 0.523809523809524, 'DDoS_UDP': 0.8465473145780052, 'DDoS_TCP': 0.7333333333333333}
b = {'DDoS_HTTP': 1.0, 'DDoS_UDP': 0.949874686716792, 'DDoS_TCP': 0.949874686716792}

# draw a graph for a and b. values represents f1-scores
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
index = np.arange(len(a))
bar_width = 0.2
opacity = 0.8

rects1 = plt.bar(index, a.values(), bar_width, alpha=opacity, color='b', label='Edge-IIoT')
rects2 = plt.bar(index + bar_width, b.values(), bar_width, alpha=opacity, color='r', label='Bot-IoT')

plt.xlabel('Attack Type')
plt.ylabel('F1-score')
plt.title('F1-score by Attack Type')
plt.xticks(index + bar_width, a.keys())
plt.legend()

plt.tight_layout()
plt.show()
```

```python
precision_benign_a = {'DDoS_HTTP': 1.0, 'DDoS_UDP': 1.0, 'DDoS_TCP': 1.0}
precision_attack_a = {'DDoS_HTTP': 0.5555555555555556, 'DDoS_UDP': 0.7692307692307693, 'DDoS_TCP': 0.6666666666666666}
recall_benign_a = {'DDoS_HTTP': 0.2, 'DDoS_UDP': 0.7, 'DDoS_TCP': 0.5}
recall_attack_a = {'DDoS_HTTP': 1.0, 'DDoS_UDP': 1.0, 'DDoS_TCP': 1.0}

precision_benign_b = {'DDoS_HTTP': 1.0, 'DDoS_UDP': 1.0, 'DDoS_TCP': 1.0}
precision_attack_b = {'DDoS_HTTP': 1.0, 'DDoS_UDP': 0.9090909090909091, 'DDoS_TCP': 0.9090909090909091}
recall_benign_b = {'DDoS_HTTP': 1.0, 'DDoS_UDP': 0.9, 'DDoS_TCP': 0.9}
recall_attack_b = {'DDoS_HTTP': 1.0, 'DDoS_UDP': 1.0, 'DDoS_TCP': 1.0}


# draw a graph with for 4 subplots for precision-benign, precision-attack, recall-benign, recall-attack
# in each graph compare a and b
fig, axs = plt.subplots(2, 2)
fig.suptitle('Precision and Recall by Attack Type')
index = np.arange(len(precision_benign_a))
bar_width = 0.2
opacity = 0.8

axs[0, 0].bar(index, precision_benign_a.values(), bar_width, alpha=opacity, color='b', label='Edge-IIoT')
axs[0, 0].bar(index + bar_width, precision_benign_b.values(), bar_width, alpha=opacity, color='r', label='Bot-IoT')
axs[0, 0].set_title('Precision Benign')
axs[0, 0].set_xticks(index + bar_width)
axs[0, 0].set_xticklabels(precision_benign_a.keys())
axs[0, 0].legend()

axs[0, 1].bar(index, precision_attack_a.values(), bar_width, alpha=opacity, color='b', label='Edge-IIoT')
axs[0, 1].bar(index + bar_width, precision_attack_b.values(), bar_width, alpha=opacity, color='r', label='Bot-IoT')
axs[0, 1].set_title('Precision Attack')
axs[0, 1].set_xticks(index + bar_width)
axs[0, 1].set_xticklabels(precision_attack_a.keys())
axs[0, 1].legend()

axs[1, 0].bar(index, recall_benign_a.values(), bar_width, alpha=opacity, color='b', label='Edge-IIoT')
axs[1, 0].bar(index + bar_width, recall_benign_b.values(), bar_width, alpha=opacity, color='r', label='Bot-IoT')
axs[1, 0].set_title('Recall Benign')
axs[1, 0].set_xticks(index + bar_width)
axs[1, 0].set_xticklabels(recall_benign_a.keys())
axs[1, 0].legend()

axs[1, 1].bar(index, recall_attack_a.values(), bar_width, alpha=opacity, color='b', label='Edge-IIoT')
axs[1, 1].bar(index + bar_width, recall_attack_b.values(), bar_width, alpha=opacity, color='r', label='Bot-IoT')
axs[1, 1].set_title('Recall Attack')
axs[1, 1].set_xticks(index + bar_width)
axs[1, 1].set_xticklabels(recall_attack_a.keys())
axs[1, 1].legend()

plt.tight_layout()
plt.show()

```
