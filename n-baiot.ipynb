```python
import pandas as pd

benign_df = pd.read_csv('data/nbaiot/5.benign.csv')
g_c_df = pd.read_csv('data/nbaiot/5.gafgyt.combo.csv')
g_j_df = pd.read_csv('data/nbaiot/5.gafgyt.junk.csv')
g_s_df = pd.read_csv('data/nbaiot/5.gafgyt.scan.csv')
g_t_df = pd.read_csv('data/nbaiot/5.gafgyt.tcp.csv')
g_u_df = pd.read_csv('data/nbaiot/5.gafgyt.udp.csv')
m_a_df = pd.read_csv('data/nbaiot/5.mirai.ack.csv')
m_sc_df = pd.read_csv('data/nbaiot/5.mirai.scan.csv')
m_sy_df = pd.read_csv('data/nbaiot/5.mirai.syn.csv')
m_u_df = pd.read_csv('data/nbaiot/5.mirai.udp.csv')
m_u_p_df = pd.read_csv('data/nbaiot/5.mirai.udpplain.csv')
```

```python
benign_df['type'] = 'benign'
m_u_df['type'] = 'mirai_udp'
g_c_df['type'] = 'gafgyt_combo'
g_j_df['type'] = 'gafgyt_junk'
g_s_df['type'] = 'gafgyt_scan'
g_t_df['type'] = 'gafgyt_tcp'
g_u_df['type'] = 'gafgyt_udp'
m_a_df['type'] = 'mirai_ack'
m_sc_df['type'] = 'mirai_scan'
m_sy_df['type'] = 'mirai_syn'
m_u_p_df['type'] = 'mirai_udpplain'
```

```python
df = pd.concat([benign_df, m_u_df, g_c_df,
                g_j_df, g_s_df, g_t_df,
                g_u_df, m_a_df, m_sc_df,
                m_sy_df, m_u_p_df],
                axis=0, sort=False, ignore_index=True)
```

```python
df["type"].value_counts()
```

```python
from matplotlib import pyplot as plt

plt.title("Class Distribution")
df.groupby("type").size().plot(kind='pie', autopct='%.2f', figsize=(20,10))
```
