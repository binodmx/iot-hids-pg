```python
import os
import pandas as pd

df_train = pd.read_csv(os.getcwd() + '/../data/unsw-nb15/CSV Files/Training and Testing Sets/UNSW_NB15_training-set.csv', low_memory=False)
df_test = pd.read_csv(os.getcwd() + '/../data/unsw-nb15/CSV Files/Training and Testing Sets/UNSW_NB15_testing-set.csv', low_memory=False)
```

```python
print("No of Featueres: ", len(df_train.columns))
print("Features: ", df_train.columns)
```

```python
print(df_train.head(1).to_json(orient='records', lines=True))
```

```python
print("No of classes: ", len(df_train['attack_cat'].unique()))
print(df_train['attack_cat'].unique())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 6))
plt.xticks(rotation=45)
# log scale

s = sns.countplot(data=df_train, x='attack_cat', hue='attack_cat')
for p in s.patches:
    s.annotate(format(p.get_height(), '.0f'), 
               (p.get_x() + p.get_width() / 2., p.get_height()), 
               ha = 'center', va = 'center', 
               xytext = (0, 9), 
               textcoords = 'offset points')
plt.title('Record Count', fontsize=20)

plt.show()
```
